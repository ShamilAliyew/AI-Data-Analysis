```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/Users/samil/Desktop/Programming/DataHandling/CVSData1/Total AI Publication Count.csv')
print(df.head())
```

```python
df.head()
```

```python
df = df.rename(columns={'Number of AI Publications':'Number of Publications'})
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Year'], df['Number of Publications'], linestyle= '-', marker ='*',color='g')
plt.title('AI Publications for Each Year')
plt.xlabel('Years')
plt.ylabel('Number of Publications')
plt.grid()
plt.tight_layout()
plt.show()
```
