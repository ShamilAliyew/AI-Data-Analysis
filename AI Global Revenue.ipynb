```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv('CVSData2/AI software market global annual revenue (2018 to 2025).csv',skiprows=1)
df.head()
```

```python
print(df['Revenue'].dtype)
```

```python

plt.figure(figsize=(10,6))
plt.bar(df['Year'] , df['Revenue'], color='orange')
plt.title('Global software revenue of AI in 2018-1025')
plt.xlabel('Years')
plt.ylabel('Revenue (Billion $)')
plt.tight_layout()
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
total_revenue=df['Revenue'].sum()
total_revenue
```

```python
plt.figure(figsize=(10,6))
colors = plt.cm.Set3.colors[:len(df['Revenue'])]
patches,texts, autotexts =plt.pie(
    df['Revenue'],
    autopct=lambda p: '{:.1f}%'.format(p),
    colors=colors,
    startangle=140 
    
)
plt.legend(patches, df['Year'], title='Years', loc='center left', bbox_to_anchor=(1, 0.5))

plt.title(f'Percentage share of AI revenue by year\nTotal Revenue is ${total_revenue} Billion')
plt.tight_layout()
plt.show()
```
