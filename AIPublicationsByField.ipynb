```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/samil/Desktop/Programming/DataHandling/CVSData1/AI Publication by Field.csv")
df.head()
```

```python
pivotDf = df.pivot(index='Year', columns='Field of Study', values='Number of AI Publications')
pivotDf.head()
```

```python
pivotDf = pivotDf.rename(columns={'Human‚Äìcomputer interaction':'Human,AI computer interaction','Machine learning':'ML','Natural language processing':'NLP'})
```

```python
plt.figure(figsize=(12,6))
for column in pivotDf.columns:
    plt.plot(pivotDf.index,pivotDf[column], marker='o',label=column, linewidth='0.8')
    
plt.title('AI Publications By Field (2010-2021)', fontsize='16')
plt.xlabel('Years',fontsize='13')
plt.ylabel('Number of Publications', fontsize='13')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
