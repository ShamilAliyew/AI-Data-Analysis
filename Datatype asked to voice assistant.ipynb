```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('CVSData2/Datatype asked to voice assistant.csv',skiprows=1)
df = df.drop('0',axis=1)
df.head()
```

```python

x = range(len(df))  
width = 0.25  

plt.figure(figsize=(12,6))

plt.bar([ i - width for i in x], df['Google Assistant'].str.rstrip('%').astype(float), width=width, label='Google Assistant', color='green')
plt.bar(x, df['Siri'].str.rstrip('%').astype(float), width=width, label='Siri', color='skyblue')
plt.bar([i + width for i in x], df['Alexa'].str.rstrip('%').astype(float), width=width, label='Alexa', color='orange')


plt.xticks(x, df['Type of question'], rotation=45)
plt.ylabel('Percentage of correct answers (%)')
plt.title('Voice Assistant Comparison of Response Accuracy')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(50,110)
plt.tight_layout()
plt.show()
```
