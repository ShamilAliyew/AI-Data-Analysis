```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv('CVSData2/Global ai software market growth.csv',skiprows=1)
df.head()
```

```python
df=df.drop('0',axis=1)
df['Year-Over-Year Growth']= df['Year-Over-Year Growth'].str.replace('%','').astype(float)/100
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Year'],df['Year-Over-Year Growth'],marker='o')
plt.xlabel('Years',fontsize='13')
plt.title('Global AI Software Market Growth',fontsize='16')
plt.xlabel('Years',fontsize='13')
plt.ylabel('Year-Over-Year Growth',fontsize='13')
plt.grid(True,axis='x',linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
                            
```
