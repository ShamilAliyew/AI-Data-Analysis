```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/samil/Desktop/Programming/DataHandling/CVSData1/Relative AI Skill Penetration-2015-2021.csv')
df.head()
```

```python
df=df.rename(columns={'Relative AI Skills Penetration - 2015-2021' :'Skills Penetration'})
df.head()
```

```python
plt.figure(figsize=(10,6))
plt.barh(df['Country'],df['Skills Penetration'],color='b')
plt.title('Relative AI Skills Penetration by Country  (2015-2021)', fontsize=14)
plt.xlabel('Relative AI Skills Penetration', fontsize = '13')
plt.ylabel('Countries',fontsize='13')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.grid(axis='x',linestyle='--',alpha= 0.5)
plt.show()
```
